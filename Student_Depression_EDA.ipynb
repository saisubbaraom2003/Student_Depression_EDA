# **About Dataset**
A student depression dataset typically contains data aimed at analyzing, understanding, and predicting depression levels among students. It may include features such as demographic information (age, gender), academic performance (grades, attendance), lifestyle habits (sleep patterns, exercise, social activities), mental health history, and responses to standardized depression scales.

These datasets are valuable for research in psychology, data science, and education to identify factors contributing to student depression and to design early intervention strategies. Ethical considerations like privacy, informed consent, and anonymization of data are crucial in working with such sensitive information.

# Dataset Description

## Columns
- **ID**: Unique identifier for each student.
- **Age**: Age of the student.
- **Gender**: Gender (e.g., Male, Female).
- **City**: Geographic region.
- **CGPA**: Grade Point Average or other academic scores.
- **Sleep Duration**: Average daily sleep duration.
- **Profession**: Profession of the individual.
- **Work Pressure**: Level of work-related stress.
- **Academic Pressure**: Level of academic stress.
- **Study Satisfaction**: Satisfaction with studies.
- **Job Satisfaction**: Satisfaction with job.
- **Dietary Habits**: Eating habits (e.g., healthy, unhealthy).
- **And much more**: Additional features or attributes.

## Target Variable
- **Depression_Status**: Binary (Yes/No).

```python
#Droping Unwanted Columns
df.drop(columns=["id","Work Pressure","Job Satisfaction"],inplace=True)
#Droping Other Professions except Stident
df=df[df["Profession"]=="Student"]
```

```python
df.head()
```

```python
print(f"{round(df['Depression'].mean()*100,2)} %  Students Getting Depression")
```

```python
#Analysing Influence Of Aspects On Depression
for i in df.keys():
  # if df[i].dtype=='O':
    if i not in ["Profession","City","Degree","Age","CGPA","Depression"]:
      ax=sns.barplot(df,x=df[i],y=df["Depression"],hue=df["Depression"],estimator=len)
      for i in ax.containers:
        ax.bar_label(i,fmt="%d")
      plt.show()
ax=sns.histplot(df,x="CGPA",hue="Depression",bins=[5,6,7,8,9,10],multiple="dodge")
for i in ax.containers:
      ax.bar_label(i,fmt="%d")
plt.show()
ax=sns.histplot(df,x="Age",
                hue="Depression",
                bins=[18,20,22,24,26,28,30,32,34],
                multiple="dodge")#Ignoring above 34 age groups due to neglisible impact of Depression on them
for i in ax.containers:
      ax.bar_label(i,fmt="%d")
plt.show()
```

```python
#City Wise Analysis
df1=df[df["Depression"]==1]
a=pd.DataFrame({"Depression = 1":df1.groupby("City")["Depression"].count()})
df2=df[df["Depression"]==0]
b=pd.DataFrame({"Depression = 0":df2.groupby("City")["Depression"].count()})
City_Wise=pd.concat([a,b],axis=1)
print("City Wise Analysis")
City_Wise
```

```python
#Degree Wise Analysis
df3=df[df["Depression"]==1]
c=pd.DataFrame({"Depression = 1":df3.groupby("Degree")["Depression"].count()})
df4=df[df["Depression"]==0]
d=pd.DataFrame({"Depression = 0":df4.groupby("Degree")["Depression"].count()})
Degree_Wise=pd.concat([c,d],axis=1)
print("Degree Wise Analysis")
Degree_Wise
```

# Key Observations on Depression Among Students

- Both boys and girls are affected by depression, but the number of boys experiencing it is comparatively higher.  
- Students with academic pressure greater than 3 are more likely to experience depression.  
- Students with a study satisfaction level below 5 are more prone to depression.  
- Sleep duration of less than 8 hours is associated with a higher likelihood of depression.  
- Healthy habits play a significant role in reducing the risk of depression.  
- Students experiencing depression are more likely to have suicidal thoughts.  
- Students who work more than 4 hours a day are more susceptible to depression.  
- Financial stress (status level greater than 3) contributes to higher depression rates.  
- Students from families with a history of mental illness, as well as those from normal family backgrounds, are both at risk of depression.  
- Students aged between 18 and 34 are more commonly affected by depression.  
- Students in Class 12 and those pursuing B.Ed programs are particularly vulnerable to depression.
